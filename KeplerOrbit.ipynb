# Kepler Orbit

```openmodelica
class KepOrbit "Elliptical Orbit"
  constant Real MU=398600.;
  Real r(start=7500.), theta(start=0.);
  Real vr(start=0.), vtheta(start=7.2938);
  output Real x, y;
equation
  der(r) = vr;
  der(theta) = vtheta/r;
  der(vr) = -MU/r^2+vtheta^2/r;
  der(vtheta) = -vr*vtheta/r;
  x = r*cos(theta);
  y = r*sin(theta);
end KepOrbit; 
```

```openmodelica
loadModelica3D();
loadString("model visualizemodel
  extends KepOrbit;
  inner ModelicaServices.Modelica3D.Controller m3d_control;
end visualizemodel;"); 
getErrorString();
simulate(visualizemodel, startTime=0.0,stopTime=6500.0); 
```
